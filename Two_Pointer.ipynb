```python
"""
https://leetcode.com/problems/implement-strstr/
Implement strStr().

Return the index of the first occurrence of needle in haystack, or -1 if needle is not part of haystack.

Clarification:

What should we return when needle is an empty string? This is a great question to ask during an interview.

For the purpose of this problem, we will return 0 when needle is an empty string. This is consistent to C's strstr() and Java's indexOf().

 """

"""
Thought Process:
Have a pointer for haystack
Have a pointer for needle
Need a boolean to know that we have found the first index
if we find that the string is not the same while comparing then 
we need to go to the position to the next of ans to ensure we check every
character.
"""
class Solution:
    def strStr(self, haystack: str, needle: str) -> int:
        if len(needle) == 0:
            return 0
        h_p = 0
        n_p = 0
        ans = 0
        found = False
        while h_p < len(haystack) and n_p < len(needle):
            if haystack[h_p] == needle[n_p]:
                if not found:
                    ans = h_p
                    found = True
                h_p += 1
                n_p += 1
                if n_p == len(needle):
                    return ans
            else:
                if found:
                    h_p = ans + 1
                else:
                    h_p += 1
                n_p = 0
                ans = 0
                found = False
        return -1
```

```python

from typing import List
"""
https://leetcode.com/problems/move-zeroes/
Given an integer array nums, move all 0's to the end of it while maintaining the relative order of the non-zero elements.

Note that you must do this in-place without making a copy of the array.
"""
"""
Thought process:
Need to swap only if the first index is 0 and second index is not zero
might need to traverse entire array in an O(n^2) fashion.
"""
class Solution:
    def moveZeroes(self, nums: List[int]) -> None:
        """
        Do not return anything, modify nums in-place instead.
        """
        if len(nums) <= 1:
            return nums
        
        i = 0
        j = i + 1
        while j < len(nums) and i < len(nums):
            if nums[i] ==  0 and nums[j] != 0:
                nums[i] = nums[j]
                nums[j] = 0
                i += 1
            j += 1
            if j == len(nums):
                i += 1
                j = i + 1
            
```

```python
"""
https://leetcode.com/problems/3sum/
Given an integer array nums, return all the triplets [nums[i], nums[j], nums[k]] such that i != j, i != k, and j != k, and nums[i] + nums[j] + nums[k] == 0.

Notice that the solution set must not contain duplicate triplets.

"""

"""
Lock in 2 numbers each time and find the third one
if not(i > 2)... because the numbers are sorted and we can detect it in this way.
"""
def threeSum(nums: List[int]) -> List[List[int]]:
    res = set()
    nums.sort() #Sorting the number!!!!!!!!
    _set = set()
    print(nums)
    for i in range(len(nums)):
        if not(i > 2 and nums[i] == nums[i-2]): #this avoids selecting 2 same numbers that were selected before
            for j in range(i+1, len(nums)):
                print(i, j, res, _set)
                sol = -(nums[i] + nums[j])
                if sol in _set:
                    res.add((nums[i], nums[j], sol))
            _set.add(nums[i])
    return list(res)
        
print(threeSum([-1,-1,1,2,-1,-4]))
```

```python
import math
def threeSumClosest(nums: List[int], target: int) -> int:
    output = math.inf
    nums.sort()
    print(nums)
    for i in range(1, len(nums)):
        for j in range(i + 1, len(nums)):
            temp_sum = nums[i] + nums[j] + nums[i-1]
            if abs(temp_sum - target) < abs(output - target):
                output = temp_sum
    return output

print(threeSumClosest([-1,2,1,-4], 1))
```
