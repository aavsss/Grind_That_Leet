```python
"""
https://leetcode.com/problems/letter-combinations-of-a-phone-number/submissions/
Given a string containing digits from 2-9 inclusive, return all possible letter combinations that the number could represent. Return the answer in any order.

A mapping of digit to letters (just like on the telephone buttons) is given below. Note that 1 does not map to any letters.
Input: digits = "23"
Output: ["ad","ae","af","bd","be","bf","cd","ce","cf"]
"""
"""
redeclare the answer variable in each time because we have to iterate over the previous one
"""
from collections import deque
from typing import List

#just pursue the idea next time

class Solution:
    def letterCombinations(self, digits: str) -> List[str]:
        if not digits:
            return []
        
        phone_map = {
            '2' : ['a','b','c'],
            '3' : ['d','e','f'],
            '4' : ['g','h','i'],
            '5' : ['j', 'k', 'l'],
            '6' : ['m','n','o'],
            '7' : ['p','q','r','s'],
            '8' : ['t','u','v'],
            '9' : ['w','x','y','z']
        }
        
        combinations = [''] #Creating a new combination array each time. 
        for digit in digits:
            letters = phone_map[digit]
            combinations = [ #this does the trick - list comprehension
                combination + letter
                for combination in combinations
                for letter in letters
            ]
            print(combinations)
        return combinations
```

```python
"""
https://leetcode.com/problems/sum-root-to-leaf-numbers/
You are given the root of a binary tree containing digits from 0 to 9 only.

Each root-to-leaf path in the tree represents a number.

For example, the root-to-leaf path 1 -> 2 -> 3 represents the number 123.
Return the total sum of all root-to-leaf numbers. Test cases are generated so that the answer will fit in a 32-bit integer.
"""

"""
Thought process:
put the main operation in the leaf condition
and think of a way to traverse to the leaf
learn how to traverse. Revise the code time and time and do revisions. 
"""
# Definition for a binary tree node.
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right
class Solution:
    def sumNumbers(self, root: TreeNode) -> int:
        sum_list = []
        self.dfs(root, sum_list, 0)
        return sum(sum_list)
    
    def dfs(self, root, sum_list, num):
        if root.left == None and root.right == None:
            num = num * 10 + root.val
            sum_list.append(num)
        else:
            if root.left:
                self.dfs(root.left, sum_list, num * 10 + root.val)
            if root.right:
                self.dfs(root.right, sum_list, num * 10 + root.val)

        
```
