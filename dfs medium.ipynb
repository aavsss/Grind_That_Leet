```python
"""
https://leetcode.com/problems/letter-combinations-of-a-phone-number/submissions/
Given a string containing digits from 2-9 inclusive, return all possible letter combinations that the number could represent. Return the answer in any order.

A mapping of digit to letters (just like on the telephone buttons) is given below. Note that 1 does not map to any letters.
Input: digits = "23"
Output: ["ad","ae","af","bd","be","bf","cd","ce","cf"]
"""
"""
redeclare the answer variable in each time because we have to iterate over the previous one
"""
from collections import deque
from typing import List

#just pursue the idea next time

class Solution:
    def letterCombinations(self, digits: str) -> List[str]:
        if not digits:
            return []
        
        phone_map = {
            '2' : ['a','b','c'],
            '3' : ['d','e','f'],
            '4' : ['g','h','i'],
            '5' : ['j', 'k', 'l'],
            '6' : ['m','n','o'],
            '7' : ['p','q','r','s'],
            '8' : ['t','u','v'],
            '9' : ['w','x','y','z']
        }
        
        combinations = [''] #Creating a new combination array each time. 
        for digit in digits:
            letters = phone_map[digit]
            combinations = [ #this does the trick - list comprehension
                combination + letter
                for combination in combinations
                for letter in letters
            ]
            print(combinations)
        return combinations
```

```python
"""
https://leetcode.com/problems/sum-root-to-leaf-numbers/
You are given the root of a binary tree containing digits from 0 to 9 only.

Each root-to-leaf path in the tree represents a number.

For example, the root-to-leaf path 1 -> 2 -> 3 represents the number 123.
Return the total sum of all root-to-leaf numbers. Test cases are generated so that the answer will fit in a 32-bit integer.
"""

"""
Thought process:
put the main operation in the leaf condition
and think of a way to traverse to the leaf
learn how to traverse. Revise the code time and time and do revisions. 
"""
# Definition for a binary tree node.
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right
class Solution:
    def sumNumbers(self, root: TreeNode) -> int:
        sum_list = []
        self.dfs(root, sum_list, 0)
        return sum(sum_list)
    
    def dfs(self, root, sum_list, num):
        if root.left == None and root.right == None:
            num = num * 10 + root.val
            sum_list.append(num)
        else:
            if root.left:
                self.dfs(root.left, sum_list, num * 10 + root.val)
            if root.right:
                self.dfs(root.right, sum_list, num * 10 + root.val)

        
```

```python
"""
https://leetcode.com/problems/path-sum-ii/submissions/
Given the root of a binary tree and an integer targetSum, return all root-to-leaf paths where each path's sum equals targetSum.

A leaf is a node with no children.
"""

"""
Thought process:
Magic happens in the leaf so just need to traverse till there
remember that in dfs, while traversing we always reach the leaf first before traversing to other direction
once elements are added into the candidate_list, make sure to pop too to remove already checked elements
"""
# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right
class Solution:
    def pathSum(self, root: TreeNode, targetSum: int) -> List[List[int]]:
        if not root:
            return []
        final_list = []
        self.dfs(root, targetSum, final_list, [])
        return final_list
        
    def dfs(self, root, targetSum, final_list, candidate_list):
        if root.left == None and root.right == None:
            candidate_list.append(root.val)
            # print(candidate_list)
            if sum(candidate_list) == targetSum:
                temp_list = candidate_list.copy()
                final_list.append(temp_list)
        else:
            if root:
                candidate_list.append(root.val)
                if root.left:
                    self.dfs(root.left, targetSum, final_list, candidate_list)
                    if len(candidate_list) > 0:
                        candidate_list.pop()
                if root.right:
                    self.dfs(root.right, targetSum, final_list, candidate_list)
                    if len(candidate_list) > 0:
                        candidate_list.pop()
                
                
        
```

```python
"""
https://leetcode.com/problems/target-sum/
You are given an integer array nums and an integer target.

You want to build an expression out of nums by adding one of the symbols '+' and '-' before each integer in nums and then concatenate all the integers.

For example, if nums = [2, 1], you can add a '+' before 2 and a '-' before 1 and concatenate them to build the expression "+2-1".
Return the number of different expressions that you can build, which evaluates to target.
"""
"""
Thought process:
dfs + memoization
use a return statement if we wish to save intermediate values
save the middle value in a dictionary
first check the middle value and then only go forward
"""
class Solution:
    def findTargetSumWays(self, nums: List[int], target: int) -> int:
        count = 0
        memo = {}
        def dfs(num, nums, i):
            nonlocal count
            if (i, num) in memo:
                return memo[(i, num)]

            if i == len(nums) - 1:
                if num == target:
                    ans = 1
                else:
                    ans = 0
                memo[(i, num)] = ans
                return ans
                
            if i + 1 < len(nums):
                add = dfs(num + nums[i+1], nums, i + 1)
                sub = dfs(num - nums[i+1], nums, i + 1)
                memo[(i, num)] = add + sub
                return memo[(i, num)]
                
        add = dfs(nums[0], nums, 0)
        sub = dfs(-nums[0], nums, 0)
        return add + sub
        
```
